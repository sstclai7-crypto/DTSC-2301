```python

import matplotlib.ticker as ticker
plt.figure(figsize=(12, 8))
df_sorted = df.sort_values("B01001_009E")
plt.barh(df_sorted["NAME"], df_sorted["B01001_009E"])
plt.xlabel("Number of 21-Year-Old Males (Per 100)", fontsize=14)
plt.ylabel("Counties", fontsize=14)
plt.title("21-Year-Old Male Population by County", fontsize=16)
ax = plt.gca()
ax.xaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))








```
