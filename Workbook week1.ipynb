```python
Things to clean:
- 'major' are messy. not standardized.
- 'hours_sleep' doesn't have a set time. we want floats.
- 'mood' is inconsistent. imporvements could be having a rating (integer) or catergorical (dropdown).
- 'grade' is empty. all are "None".
- Some columns are string when they should be numerical (ints or floats): Coffee, hours_sleep, id, and minutes_study.
- Weird spacing.
```

```python
# 'x' is a single string we are cleaning
def remove_extra_whitespace(x):
    if x is None:
        return None
    
    x = str(x).strip()
    return x 
```

```python
#remove_extra_whitespace(raw[3]['minutes_study'])
for row in raw:
    row['minutes_study'] = remove_extra_whitespace(row['minutes_study'])
```

```python
raw
```

```python
# Make 'major' standardized: Uppercase and no extra spaces
def capitalize_major(major):
    if major is None:
        return None
    
    major = str(major).strip().upper()
    return major

```

```python
#Capitalize_major(raw[2]['major'])
for row in raw:
    row['major'] = capitalize_major(row['major'])
```
